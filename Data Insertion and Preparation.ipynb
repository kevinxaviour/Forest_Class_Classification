## Data Understaning

- Elevation(meters) - the height of the land surface above sea level
- Aspect(degrees) - compass direction that a slope faces, typically measured in degrees, 
                    from 0degree to 360degree, where 0 degree is due north and the values increase clockwise.
- Slope(degrees) -  the steepness or incline of the terrain. is a crucial factor because it affects factors like soil erosion, 
                accessibility for logging, the types of machinery that can be used, and the forest's growth and resilience.
- Horizontal_Distance_To_Hydrology(meters) -  from a specific plot or cell of land to the nearest surface water feature.
- Vertical_Distance_To_Hydrology(meters) - Vertical distance to nearest surface water
- Horizontal_Distance_To_Roadways(meters) - Horizontal distance to nearest roadway 
- Horizontal_Distance_To_Fire_Points(meters)- Horizontal distance to nearest wildfire ignition point
- Hillshade_9am(Num) - ranging from 0 to 255, that represents how much direct sunlight a 
                    specific area of the forest would receive at that precise time during summer. 
- Hillshade_Noon(Num)- 
- Hillshade_3pm (Num)-
- Wilderness_Area_1 -binary indicator (0 or 1) that specifies whether a 30m x 30m forest patch is located within the Rawah Wilderness Area
- soil type - type of soil (One hot encoded)
- Cover_Type - the specific type of dominant tree species or forest classification for a given area
    - Spruce/Fir: A mixture dominated by spruce and fir trees.
    - Lodgepole Pine: Areas where lodgepole pine is the primary tree species.
    - Ponderosa Pine: Forests with Ponderosa pine as the dominant tree.
    - Cottonwood/Willow: Areas with a mix of cottonwood and willow trees.
    - Aspen: Forests where aspen is the main tree.
    - Douglas-fir: Areas with Douglas-fir as the dominant tree species.
    - Krummholz: A term for trees that are twisted and stunted due to harsh winds and cold conditions, often found at high altitudes.

### Decoding back to original columns from one hot encoded columns

```python
one_hot_cols1=['Wilderness_Area_1','Wilderness_Area_2', 'Wilderness_Area_3', 'Wilderness_Area_4']
one_hot_cols2=['Soil_Type_1', 'Soil_Type_2', 'Soil_Type_3', 'Soil_Type_4',
       'Soil_Type_5', 'Soil_Type_6', 'Soil_Type_7', 'Soil_Type_8',
       'Soil_Type_9', 'Soil_Type_10', 'Soil_Type_11', 'Soil_Type_12',
       'Soil_Type_13', 'Soil_Type_14', 'Soil_Type_15', 'Soil_Type_16',
       'Soil_Type_17', 'Soil_Type_18', 'Soil_Type_19', 'Soil_Type_20',
       'Soil_Type_21', 'Soil_Type_22', 'Soil_Type_23', 'Soil_Type_24',
       'Soil_Type_25', 'Soil_Type_26', 'Soil_Type_27', 'Soil_Type_28',
       'Soil_Type_29', 'Soil_Type_30', 'Soil_Type_31', 'Soil_Type_32',
       'Soil_Type_33', 'Soil_Type_34', 'Soil_Type_35', 'Soil_Type_36',
       'Soil_Type_37', 'Soil_Type_38', 'Soil_Type_39', 'Soil_Type_40']
df['Wilderness_Area']=df[one_hot_cols1].idxmax(axis=1)
df['Soil_Type']=df[one_hot_cols2].idxmax(axis=1)
df['Wilderness_Area']=df['Wilderness_Area'].str.split('_').str[2].astype(int)
df['Soil_Type']=df['Soil_Type'].str.split('_').str[2].astype(int)
```

```python
df.drop(columns=['Soil_Type_1', 'Soil_Type_2', 'Soil_Type_3', 'Soil_Type_4',
       'Soil_Type_5', 'Soil_Type_6', 'Soil_Type_7', 'Soil_Type_8',
       'Soil_Type_9', 'Soil_Type_10', 'Soil_Type_11', 'Soil_Type_12',
       'Soil_Type_13', 'Soil_Type_14', 'Soil_Type_15', 'Soil_Type_16',
       'Soil_Type_17', 'Soil_Type_18', 'Soil_Type_19', 'Soil_Type_20',
       'Soil_Type_21', 'Soil_Type_22', 'Soil_Type_23', 'Soil_Type_24',
       'Soil_Type_25', 'Soil_Type_26', 'Soil_Type_27', 'Soil_Type_28',
       'Soil_Type_29', 'Soil_Type_30', 'Soil_Type_31', 'Soil_Type_32',
       'Soil_Type_33', 'Soil_Type_34', 'Soil_Type_35', 'Soil_Type_36',
       'Soil_Type_37', 'Soil_Type_38', 'Soil_Type_39', 'Soil_Type_40',
       'Wilderness_Area_1','Wilderness_Area_2', 'Wilderness_Area_3', 'Wilderness_Area_4'],inplace=True)
df.columns
```

### Converting Back to One Hot Encoded Values


```python
one_hot_encoder_wild=OneHotEncoder()
wilderness_encoder=one_hot_encoder_wild.fit_transform(df[['Wilderness_Area','Soil_Type']]).toarray().astype(int)
wilderness_encoder
```

```python
one_hot_encoder_wild.get_feature_names_out(['Wilderness_Area','Soil_Type'])
```

```python
wilderness_encoded=pd.DataFrame(wilderness_encoder,columns=one_hot_encoder_wild.get_feature_names_out(['Wilderness_Area','Soil_Type']))
wilderness_encoded.columns
```

```python
df=pd.concat([df.drop(['Wilderness_Area','Soil_Type'],axis=1),wilderness_encoded],axis=1)
df.head()
```

```python
#Saving onehotenoder for the model into a pickle file
with open('ohe_wildernessandsoil.pkl','wb') as file:
    pickle.dump(one_hot_encoder_wild,file)
```

```python
# Checking If One hot encoder works
with open('ohe_wildernessandsoil.pkl', 'rb') as file:
    loaded_encoder = pickle.load(file)
new_df = pd.DataFrame({
    'Wilderness_Area': [1, 3],
    'Soil_Type': [11, 2]
})

encoded = loaded_encoder.transform(new_df[['Wilderness_Area', 'Soil_Type']]).toarray().astype(int)
cols = loaded_encoder.get_feature_names_out(['Wilderness_Area', 'Soil_Type'])
encoded_df = pd.DataFrame(encoded, columns=cols)
encoded_df
```
